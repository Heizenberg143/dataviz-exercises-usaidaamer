```python
# Step 1 — Imports
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Step 2 — Load dataset
df = px.data.gapminder()
```

```python
# Step 3 — Choose continent and years
continent_choice = "Asia"   # Change if you want (NOT Africa)
years = [2002, 2007]
```

```python
# Step 4 — Filter data
df_filtered = df[
    (df["continent"] == continent_choice) &
    (df["year"].isin(years))
].copy()

df_filtered["year"] = df_filtered["year"].astype(str)
```

```python
# Step 5 — Plot with Plotly Express (FIRST)

fig = px.scatter(
    df_filtered,
    x="gdpPercap",
    y="lifeExp",
    color="year",
    hover_name="country",
    log_x=True,
    color_discrete_map={
        "2002": "lightgrey",
        "2007": "blue"
    },
    title="Asia shows clear upward movement: higher income links to longer life (2002 → 2007)",
    labels={
        "gdpPercap": "GDP per capita (log scale)",
        "lifeExp": "Life Expectancy"
    }
)
```

```python
# Step 6 — Add movement arrows (Graph Objects AFTER px)

df_wide = df_filtered.pivot(
    index="country",
    columns="year",
    values=["gdpPercap", "lifeExp"]
).dropna()

for country in df_wide.index:
    x0 = df_wide.loc[country, ("gdpPercap", "2002")]
    y0 = df_wide.loc[country, ("lifeExp", "2002")]
    x1 = df_wide.loc[country, ("gdpPercap", "2007")]
    y1 = df_wide.loc[country, ("lifeExp", "2007")]

    fig.add_annotation(
        x=x1, y=y1,
        ax=x0, ay=y0,
        xref="x", yref="y",
        axref="x", ayref="y",
        showarrow=True,
        arrowhead=2,
        arrowwidth=1,
        opacity=0.4
    )
```

```python
# Step 1 — Filter 2007 only (reuse df from before)

df_2007 = df[df["year"] == 2007].copy()
```

```python
# Step 2 — Highlight one continent

highlight_continent = "Asia"  # change if you want

df_2007["highlight"] = df_2007["continent"].apply(
    lambda x: highlight_continent if x == highlight_continent else "Other"
)
```

```python
# Step 3 — Plot (Plotly Express FIRST)

fig = px.scatter(
    df_2007,
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    color="highlight",
    hover_name="country",
    log_x=True,
    size_max=60,
    color_discrete_map={
        highlight_continent: "blue",
        "Other": "lightgrey"
    },
    title=f"{highlight_continent} dominates in population scale, but income gaps still drive life expectancy",
    labels={
        "gdpPercap": "GDP per capita (log scale)",
        "lifeExp": "Life Expectancy"
    }
)
```

```python
# Step 4 — Add annotation (Graph Objects AFTER px)

fig.add_annotation(
    x=4000,
    y=65,
    text=f"{highlight_continent} contains many mid-income countries,<br>but very large populations amplify its global impact.",
    showarrow=True,
    arrowhead=2,
    ax=100,
    ay=-80,
    bgcolor="white",
    bordercolor="black"
)
```

```python
# Step 5 — Clean layout

fig.update_traces(
    marker=dict(opacity=0.75, line=dict(width=0.5, color="white"))
)

fig.update_layout(
    template="plotly_white",
    width=900,
    height=600,
    xaxis=dict(type="log", range=[2, 5]),
    yaxis=dict(range=[35, 85]),
    legend_title_text="Group"
)

fig.show()
```
